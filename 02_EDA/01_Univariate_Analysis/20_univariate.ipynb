### Catagorical Features

#### a.Countplot

```python
sns.countplot(df, x="Pclass")
```

```python
sns.countplot(df, x="Pclass", hue="Survived")
```

```python
sns.countplot(df, x="Pclass", hue="Survived", stat="percent")
```

## 2.Numerical Data

### a.Histogram

```python
plt.hist(df['Age'])
```

###  Histogram & KDE  Replacement for `distplot`

**What:**  
Visualizes the **distribution of a continuous variable** using a histogram (bars) and KDE (smoothed probability curve).

**Why Use It:**  
- Understand **data shape** (e.g., normal, skewed).  
- Detect **outliers**, **peaks**, or **gaps**.  
- Decide on **scaling or transformation** needs.

```python
sns.histplot(data=df, x='Age', kde=True, bins=30)
```

### Showing Density Curve

```python
sns.kdeplot(df["Age"])
```

###  BoxPlot  Quick Guide

**What:**  
A BoxPlot visualizes the **distribution of a numerical variable** through **quartiles**, showing:
- **Median (Q2)**
- **25th & 75th percentiles (Q1 & Q3)**
- **Outliers** (as individual points)
- **IQR-based whiskers**

**Why Use It:**  
- Quickly detect **skewness**, **spread**, and **outliers**  
- Compare distributions **across categories**
- Useful for **feature analysis** and **outlier handling**

```python
sns.boxplot(df['Fare'])
```
